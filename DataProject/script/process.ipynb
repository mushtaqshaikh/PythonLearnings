```python
# %pip install matplotlib - To install matplotlib when it doesnt work
# %pip install seaborn
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/SuperStore_Sales_Dataset.csv')
```

```python
df_filtered = df.drop(df.columns[0], axis=1)
```

```python
df_filtered['Month'] = pd.DatetimeIndex(df_filtered['Order Date']).month

```

```python
summary = df_filtered.describe()
print(summary)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Sales', data=df_filtered)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales over time')
plt.show()
```
